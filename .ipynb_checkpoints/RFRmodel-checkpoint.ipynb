```python
import numpy as np
import pandas as pd
from pandas import DataFrame as DF, Series as Se
from sklearn.externals import joblib
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.externals import joblib
```

```python
df_train = pd.read_csv('data/fix_train.csv')
df_test = pd.read_csv('data/fix_test.csv')
time = df_train['playtime_forever']
```

```python
rfr = joblib.load('model/rfr.pkl')
```

```python
###############################################
#This is the training part here!              #
#Please skip from here if directly load model!#
###############################################
```

```python
rfr = RandomForestRegressor()
cv_params = {'n_estimators' : range(30, 80, 5),
             'max_features' : [0.1,0.12,0.14,0.16,0.18,0.2,0.22]}
grid = GridSearchCV(estimator=rfr, param_grid=cv_params, n_jobs=-1, cv=5,verbose=2,scoring='neg_mean_squared_error')
```

```python
# grid search for best parameters
grid.fit(df_train[df_test.columns], time)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
rfr = RandomForestRegressor(**grid.best_params_)
rfr.fit(df_train[df_test.columns], time)
```

```python
#########################################
#Start from here if directly load model!#
#########################################
```

```python
test_pred = rfr.predict(df_test)
df_pred = DF((test_pred))
df_pred.columns = ['playtime_forever']
```

```python
# dealing with possible outliers
tmp = []
for x in df_pred['playtime_forever']:
    if x < 0.95:
        tmp.append(0)
    elif x > 20:
        tmp.append(2.5 * x)
    elif x > 10:
        tmp.append(1.5 * x)
    else:
        tmp.append(x)
df_pred['playtime_forever'] = tmp
```

```python
df_pred['id'] = range(test_pred.shape[0])
df_pred.to_csv('RFR_res.csv', index=False)
```

```python
# saving rfr model
joblib.dump(rfr, 'model/rfr.pkl')
```
